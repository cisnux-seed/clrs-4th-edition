Loop Invariant:
At the start of each iteration of the loop in SUM-ARRAY, let's define a loop invariant as follows:

The variable 'sum' contains the sum of the elements A[1:i], where i represents the current index being processed within the loop.

Initialization:
Before the loop begins (when i = 1), the loop invariant holds true because the sum is initialized to 0, and no elements from A[1:i] are added yet. Therefore, the sum is indeed the sum of zero elements, which is 0, meeting the loop invariant.

Maintenance:
During each iteration of the loop, the loop invariant is maintained. As the loop progresses, the algorithm adds the current element A[i] to the 'sum'. At any iteration with index i, the 'sum' accurately represents the sum of elements A[1:i]. When moving to the next index (i+1), the loop continues to maintain the loop invariant by adding A[i+1] to the 'sum'.

Termination:
Upon completion of the loop (when i = n), the loop invariant still holds. The loop ends after processing all elements in A[1:n]. At this point, the 'sum' contains the sum of all elements A[1:n] as per the loop invariant. The loop terminates once i equals n+1.

````python
def add_binary_integers(a, b) -> list:
    n = max(len(a), len(b))
    c = [0 for _ in range(n + 1)]
    carry = 0

    for i in range(0, len(c)):
        # a_i = 0
        # b_i = 0
        # if i < len(a):
        #     a_i = a[i]
        # if i < len(b):
        #     b_i = b[i]
        a_i = 0 if i >= len(a) else a[i] 
        b_i = 0 if i >= len(b) else b[i] 
        c[i] = (carry + a_i + b_i) % 2
        carry = (carry + a_i + b_i) // 2
    return c
````

````python
def decimal_to_binary(x) -> list:
    carry = x
    bin = []

    while carry > 0:
        bin.append(carry % 2)
        carry = carry // 2
    return bin
````

````python
def binary_to_decimal(a) -> int:
    decimal = 0
    for i in range(0, len(a)):
        decimal += a[i] * 2**i
    return decimal
````

````python
def descending_insertion_sort(a) -> int:
    for i in range(1, len(a)):
        key = a[i]
        # insert a[i] into the sorted subarray a[1:i-1]
        j = i - 1
        while j > -1 and a[j] < key:
            a[j + 1] = a[j]
            j -= 1  # -> j = j-1
        # If all the cards in your left hand have values greater than the card in your right hand,
        # then place this card as the leftmost card in your left hand.
        a[j + 1] = key
````

````python
def linear_search(a, x) -> int:
    for i in range(0, len(a)):
        if a[i] == x:
            return i
    return None
````
